## Mutable parameter default value

Avoid using default mutable arguments.

Argument defaults are defined when the function is defined, not when it is exercised.

```python
# wrong
def append(n, l=[]):
    l.append(n)
    return l

print(append(0)) # [0]
print(append(1)) # [0, 1]  Oops!
```

```python
# improved
def append(n, l=None):
    if l is None:
        l = []
    l.append(n)
    return l

print(append(0)) # [0]
print(append(1)) # [1]
```
